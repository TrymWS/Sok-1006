### Analyse av en merverdiavgift ("ad valorem tax")

Her skal vi se på virkningen av å ilegge en vare en skatt i henhold til dens verdi (merverdiavgift). Betrakt fremstillingen av markedet for oppdrettslaks i læreboka s. 101-104. Etterspørsel er gitt ved


\begin{equation}
   x^{D}= 500 -3.2p_k\tag{1}
\end{equation}

og tilbud:


\begin{equation}
   x^{S}= -100 + 4.3p_s\tag{2}
\end{equation}

Kjøpere betaler $p_k$ og selgere får $p_s$. Prisen som selgere får er det som konsumenter betaler minus mva $tp_k$, noe som betyr at $p_s=(1-t)p_k$. Her er $1>t\geq 0$ avgiftssatsen. Vi kan skrive (2) også som en funksjon på prisen som konsumenten betaler:

\begin{equation}
   x^{S}= -100 + 4.3 p_k(1-t)\tag{3}
\end{equation}



```python
# importer pakkene
import numpy as np
from matplotlib import pyplot as plt
import sympy as sp
```

```python
# Etterspørsel i likning (1)

def x_d(p):
    return 500 - 3.2*p

#Tilbud i likning (3)

def x_s(p,t):
    return -100+4.3*p*(1-t)
```

```python
# vi lager en basisfigur

p_num=np.linspace(0.1,100,100)

def create_ax():
                fig, ax = plt.subplots()
                ax.set_ylabel('Pris per kilo', loc='top')
                ax.set_xlabel('Tonn per uke', loc='right')
                ax.set(xlim=(75,280))
                ax.set(ylim=(40,100))
                ax.spines['top'].set_color('none')
                ax.spines['right'].set_color('none')
                
                return fig, ax

fig, ax = create_ax()

# plott funksjonene

ax.plot(x_d(p_num),p_num, label='Etterspørsel')

ax.plot(x_s(p_num, 0),p_num, label='Tilbud')

ax.legend(bbox_to_anchor=(1.5,0.6));
```

```python
def x_s_25(p,t):
    return -100+4.3*(p/1.25)

def x_s_250(p,t):
    return -100+4.3*p*(1-0.25)


# vi lager en basisfigur

p_num=np.linspace(0.1,100,100)

def create_ax():
                fig, ax = plt.subplots()
                ax.set_ylabel('Pris per kilo', loc='top')
                ax.set_xlabel('Tonn per uke', loc='right')
                ax.set(xlim=(75,280))
                ax.set(ylim=(40,100))
                ax.spines['top'].set_color('none')
                ax.spines['right'].set_color('none')
                
                return fig, ax

fig, ax = create_ax()

# plott funksjonene

ax.plot(x_d(p_num),p_num, label='Etterspørsel')

ax.plot(x_s(p_num, 0),p_num, label='Tilbud')
ax.plot(x_s_25(p_num, 0),p_num, label='Tilbud_MVA', color='lightblue')
ax.plot(x_s_250(p_num, 0),p_num, label='Tilbud_MVA2', color='lightgreen')

ax.legend(bbox_to_anchor=(1.5,0.6));
```

Er ikke helt sikker på hva det spørres etter, siden (1-t) burde bli (1-0.25), men MVA utgjør 1/5 av prisen, og det burde etter min forståelse spørres om (p/1.25)

Tilbudsfunksjonen med MVA har en brattere helning enn uten MVA fordi prisen, totalt sett, blir høyere. 
Tilbudet er også mer prissensitivt enn etterspørsel.

2. Beregn likevektspris og -kvantum med og uten mva, og vis disse i figuren.

```python
# Fortsetter med (p/1.25)

def x_d_250(p):
    return 500 - 3.2*p*(1+0.25)

# vi lager en basisfigur

p_num=np.linspace(0.1,200,200)

def create_ax():
                fig, ax = plt.subplots()
                ax.set_ylabel('Pris per kilo', loc='top')
                ax.set_xlabel('Tonn per uke', loc='right')
                ax.set(xlim=(100,300))
                ax.set(ylim=(70,100))
                ax.spines['top'].set_color('none')
                ax.spines['right'].set_color('none')
                
                return fig, ax

fig, ax = create_ax()

# plott funksjonene

ax.plot(x_d(p_num),p_num, label='Etterspørsel')
ax.plot(x_d_250(p_num),p_num, label='Etterspørsel_MVA2', color='lightgreen')

ax.plot(x_s(p_num, 0),p_num, label='Tilbud')
ax.plot(x_s_25(p_num, 0),p_num, label='Tilbud_MVA2', color='lightgreen')

ax.vlines(211, 70, 100, color='lightblue', linestyle='--')
ax.hlines(90.4, 100, 211, color='lightblue', linestyle='--')
ax.hlines(72.25, 100, 211, color='lightblue', linestyle='--')
ax.hlines(80, 100, 244, color='black', linestyle='--')
ax.vlines(244, 70, 80, color='black', linestyle='--')

ax.legend(bbox_to_anchor=(1.5,0.6));
```

Likevektspris og kvantum uten MVA er henholdsvist 80 og 244

Likevektspris og kvantum med MVA er henholdsvist 90.4 og 211

3. Beregn og vis inntekten til staten, og andelen som betales av konsumenter og produsenter.

```python
# vi lager en basisfigur

p_num=np.linspace(0.1,200,200)

def create_ax():
                fig, ax = plt.subplots()
                ax.set_ylabel('Pris per kilo', loc='top')
                ax.set_xlabel('Tonn per uke', loc='right')
                ax.set(xlim=(100,300))
                ax.set(ylim=(70,100))
                ax.spines['top'].set_color('none')
                ax.spines['right'].set_color('none')
                
                return fig, ax

fig, ax = create_ax()

# plott funksjonene

ax.plot(x_d(p_num),p_num, label='Etterspørsel')
ax.plot(x_d_250(p_num),p_num, label='Etterspørsel_MVA2', color='lightgreen')

ax.plot(x_s(p_num, 0),p_num, label='Tilbud')
ax.plot(x_s_25(p_num, 0),p_num, label='Tilbud_MVA2', color='lightgreen')

ax.vlines(211, 70, 100, color='lightblue', linestyle='--')
ax.hlines(90.4, 100, 211, color='lightblue', linestyle='--')
ax.hlines(72.25, 100, 211, color='lightblue', linestyle='--')
ax.hlines(80, 100, 244, color='black', linestyle='--')
ax.vlines(244, 70, 80, color='black', linestyle='--')

ax.fill_between(x_d(p_num), 90.4, 80, where=(x_d(p_num) < 211), color='lightblue', alpha=0.5)
ax.fill_between(x_d(p_num), 80, 72.25, where=(x_d(p_num) < 211), color='lightgreen', alpha=0.5)

ax.legend(bbox_to_anchor=(1.5,0.6));
```

Det fargede området er inntekter til staten, altså prisen 90.4 - 72.25 = 18.15

Det som betales av konsumenter er 90.4 - 80 = 10.4

Det som betales av produsenter er 80 - 72.25 = 7.75

4. Hva er fordeler/ulemper av en mva og en stykkprisavgift?

Fordelen med MAV er at det er enklere å kalkulere, mens ulemper kan være at det blir for generelt for visse goder/tjenester.

Fordelen med stykkprisavgift er at det kan spesifiseres mer basert på ønskede/uønskede goder/tjenester, mens ulemper kan være at det er vanskeligere eller mer omfattende å sette forskjellige stykkprisavgifter for forskjellige goder/tjenester, som fører til høyere administrasjonskostnader.
